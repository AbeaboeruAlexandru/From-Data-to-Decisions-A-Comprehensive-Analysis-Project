```python
from pandas import read_csv
from google.colab import drive
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import BallTree
from sklearn.neighbors import KDTree
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.preprocessing import MinMaxScaler
import time
import os
```

```python
drive.mount('/content/drive')
path="/content/drive/MyDrive/Pulsars.csv"
dataset=read_csv(path)
dataset.head()
print(dataset.shape)
print(dataset.head(20))
```

```python
array = dataset.values
X_train = array[0:1700, 0:8]
X_validation = array[0:17898, 0:8]
Y_train = array[0:1700, 8]
Y_validation = array[0:17898, 8]

minmaxscale = MinMaxScaler().fit(X_train)
X_train = minmaxscale.transform(X_train)
X_validation = minmaxscale.transform(X_validation)
```

```python
#Cautarea liniara
model = KNeighborsClassifier(n_neighbors=41)
kfold = StratifiedKFold(n_splits=10, random_state=1, shuffle=True)

start_time = time.time()
cv_results = cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy')
end_time = time.time()

execution_time = end_time - start_time

print('%s (Cautare Liniara): Accuracy - %f (%f)' % ("kNN", cv_results.mean(), cv_results.std()))
print('%s (Cautare Liniara): Execution Time - %f seconds\n' % ("kNN", execution_time))


#k-D Tree
model = KNeighborsClassifier(n_neighbors=41, algorithm='kd_tree')
kfold = StratifiedKFold(n_splits=10, random_state=1, shuffle=True)

start_time = time.time()
cv_results = cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy')
end_time = time.time()

execution_time = end_time - start_time

print('%s (k-D Tree): Accuracy - %f (%f)' % ("kNN", cv_results.mean(), cv_results.std()))
print('%s (k-D Tree): Execution Time - %f seconds\n' % ("kNN", execution_time))
```

```python
model.fit(X_train, Y_train)
predictions = model.predict(X_validation)

print(accuracy_score(Y_validation, predictions))
print(confusion_matrix(Y_validation, predictions))
print(classification_report(Y_validation, predictions))
```
